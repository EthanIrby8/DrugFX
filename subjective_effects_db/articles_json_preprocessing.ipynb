```python
import pandas as pd
import json
```

```python
articles = pd.read_json('articles.json', lines=True)
articles
```

# Body Column 

```python
df1 = pd.DataFrame(articles['body'].values.tolist())
df1
```

```python
test1 = df1['raw'][1]

string1 = ['[p]', '[b]', '[i]', '[/n]', '[/i]', '[/li]\n  [li]']
for char in string1:
    if char in test1:
        test1 = test1.replace(char, ""+ char)
print(test1)
```

```python
parsed = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in df1.parsed.items()]))

parsed_df = pd.json_normalize(json.loads(parsed.to_json(orient='records')))
parsed_df.head()
```

```python
children_df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in parsed_df['0.children'].items()]))
# children_df.head()

children_df = pd.json_normalize(parsed_df['0.children'], )
```

```python
children_series = parsed_df['0.children'].dropna(inplace=False)
children_series.values
```

# The rows with descriptions of experience are within the integer labeled columns 

```python
bn = pd.DataFrame(children_series.values.tolist())
asdf = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in bn.items()]))
df4= pd.json_normalize(json.loads(asdf.to_json(orient='records')))
df4
```

# Created Column 

```python
created_dates_df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in articles['created'].items()])).T
created_dates_df = created_dates_df.rename(columns={'$date': 'date created'})


for row in created_dates_df['date created']:  
    created_dates_df['date created']= pd.to_datetime(created_dates_df['date created']).dt.date
created_dates_df
```

# Publication Date column

```python
publication_date_df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in articles['publication_date'].items()])).T
publication_date_df = publication_date_df.drop(columns=0)
publication_date_df = publication_date_df.rename(columns={'$date': 'publication date'})


for row in publication_date_df['publication date']:  
    publication_date_df['publication date']= pd.to_datetime(publication_date_df['publication date']).dt.date
publication_date_df
```
